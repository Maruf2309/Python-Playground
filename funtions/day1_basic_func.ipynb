## 1.1 Real-Life Example (Sales Calculation)

````python
def calculate_sales(qty, price):
    
    total = qty * price
    return total      # Return space var no operator

sales = calculate_sales(50, 100)
print(sales)
````

## Task 2- Vat Calculation

```bash
vat = price x (vat_rate/100)
```

````python
def calculate_vat(price, vat_rate):
    vat_amount = price * (vat_rate/100)
    final_price = price + vat_amount

    # Return Block
    return vat_amount, final_price

product_price = 5000        # Assume
vat_rate = 15               # as percentage, no decimal proint

vat, total = calculate_vat(product_price, vat_rate)  # must pass the var
print("VAT Amount:", vat)
print("Final Price:", total)

````

Advanced VAT Example (Invoice Style)

````python
def invoice(price, qty, vat_rate):
    # Operations
    subtotal = price * qty
    vat = subtotal * (vat_rate/100)
    grand_total = subtotal + vat

    # Return - What we expect
    return subtotal, vat, grand_total

# Variable - Var name may differ from agur
subtotal, vat, total = invoice(1200, 10, 15)

# Display
print("Subtotal:", subtotal)
print("VAT:", vat)
print("Grand Total:", total)
````

## 2. Salary Calculation Function

```bash
Basic Salary Formula:
Net Salary = Basic + Allowance + Bonus - Deduction
```

Real-Life Salary Calculation Code

````python
def calculate_salary(basic, allowance, bonus, deduction):

    gross_salary = basic + allowance + bonus
    net_salary = gross_salary - deduction

    return gross_salary, net_salary


gross, net = calculate_salary(
    basic=50000,
    allowance=10000,
    bonus=5000,
    deduction=3000
)

print("Gross Salary:", gross)
print("Net Salary:", net)
````

Advanced Salary System Example

````python
def employee_salary(name, basic, overtime_hours):

    overtime_rate = 500
    overtime_pay = overtime_hours * overtime_rate

    gross_salary = basic + overtime_pay
    tax = gross_salary * 0.05
    net_salary = gross_salary - tax

    print("Employee:", name)
    print("Gross Salary:", gross_salary)
    print("Tax:", tax)
    print("Net Salary:", net_salary)


employee_salary("Karim", 50000, 12)
````
